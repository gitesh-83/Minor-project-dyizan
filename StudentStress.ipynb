```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("StressLevelDataset.csv")
```

```python
# 1. Histogram (Countplot) of stress levels
plt.figure(figsize=(6,4))
sns.countplot(x="stress_level", data=df, palette="viridis")
plt.title("Distribution of Stress Levels")
plt.xlabel("Stress Level")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Correlation Heatmap
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", annot=False, cbar=True)
plt.title("Correlation Heatmap of Stress Dataset")
plt.show()
```

```python
# 3. Boxplot of anxiety level vs stress level
plt.figure(figsize=(6,4))
sns.boxplot(x="stress_level", y="anxiety_level", data=df, palette="Set2")
plt.title("Anxiety Level across Stress Levels")
plt.xlabel("Stress Level")
plt.ylabel("Anxiety Level")
plt.show()
```

```python
#bar 
plt.figure(figsize=(6,4))
sns.barplot(x="stress_level", y="depression", data=df, palette="mako")
plt.title("Average Depression by Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Average Depression Score")
plt.show()
```

```python
# 3. Violin plot: Sleep quality vs stress level
numeric_cols = ["anxiety_level", "depression", "sleep_quality", "stress_level"]
df_numeric = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
plt.figure(figsize=(6,4))
sns.violinplot(x="stress_level", y="sleep_quality", data=df_numeric, palette="coolwarm")
plt.title("Sleep Quality Distribution across Stress Levels")
plt.xlabel("Stress Level")
plt.ylabel("Sleep Quality")
plt.show()
```

```python
# 4. Line plot: Anxiety vs Depression
plt.figure(figsize=(6,4))
sns.lineplot(x="anxiety_level", y="depression", data=df_numeric, marker="o", color="purple")
plt.title("Trend between Anxiety and Depression")
plt.xlabel("Anxiety Level")
plt.ylabel("Depression")
plt.show()
```
